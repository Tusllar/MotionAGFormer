```python
import h5py

def print_h5_structure(h5file, indent=0):
    """
    In cấu trúc của file HDF5.
    """
    for key in h5file.keys():
        item = h5file[key]
        print("    " * indent + f"{key}: {type(item)}")
        if isinstance(item, h5py.Group):
            # Nếu là group, duyệt vào bên trong
            print_h5_structure(item, indent + 1)
        elif isinstance(item, h5py.Dataset):
            # Nếu là dataset, in thêm shape và dtype
            print("    " * (indent + 1) + f"Shape: {item.shape}, Dtype: {item.dtype}")

# Đường dẫn tới file HDF5
# h5_path = r"D:\STEP\classifier_stgcn_real_only\Data\labels.h5"
h5_path = r'D:\MotionAGFormer1\demo\labels_my.h5'

with h5py.File(h5_path, "r") as h5f:
    print_h5_structure(h5f)
```
